## Teoria

Se a resistência do ar for proporcional ao módulo da velocidade do projétil, a força resultante será dada por:

$$
\vec{F}_{\text{res}} = -mg\hat{y} - b\vec{v}.
$$

Definindo $k = b/m$, as componentes da aceleração do projétil podem ser escritas como:

$$
\left\{
\begin{align*}
a_x &= -kv_x\\
a_y &= -g-kv_y.
\end{align*}
\right.
$$

A trajetória analítica do projétil é dada pela seguinte função:

$$
y = y_0 + \frac{g}{k^2} \ln\left[1-\frac{k}{v_{x_0}}(x-x_0)\right] + \left(\frac{g}{kv_{x_0}} +\frac{v_{y_0}}{v_{x_0}}\right) (x-x_0).
$$

## Implementação

```python
# Carregando bibliotecas que serão usadas
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def aceleracao(x, y, vx, vy):
    """
    Função que calcula a aceleração do projétil.
    """
    ax = -k*vx
    ay = -g-k*vy
    return ax, ay
```

```python
def func_analitica(x):
    """
    Função que calcula a trajetória analítica do projétil.
    """
    C = g/k**2*np.log(1-k/vx0*(x-x0))
    D = (g/(k*vx0) + vy0/vx0)*(x-x0)
    return y0 + C + D
```

```python
# Métodos numéricos

def euler(x, y, vx, vy):
    """
    Atualiza a posição e a velocidade do projétil com o método de Euler
    """
    ax, ay = aceleracao(x, y, vx, vy)
    vx_novo = vx + ax*dt 
    vy_novo = vy + ay*dt
    x_novo = x + vx*dt
    y_novo = y + vy*dt
    return x_novo, y_novo, vx_novo, vy_novo

def euler_cromer(x, y, vx, vy):
    """
    Atualiza a posição e a velocidade do projétil com o método de Euler-Cromer.
    """
    ax, ay = aceleracao(x, y, vx, vy)
    vx_novo = vx + ax*dt 
    vy_novo = vy + ay*dt
    x_novo = x + vx_novo*dt
    y_novo = y + vy_novo*dt
    return x_novo, y_novo, vx_novo, vy_novo

def heun(x, y, vx, vy):
    """
    Atualiza a posição e a velocidade do projétil com o método de Heun.
    """
    ax1, ay1 = aceleracao(x, y, vx, vy)
    vx1 = vx + ax1*dt
    vy1 = vy + ay1*dt
    x1 = x + vx*dt
    y1 = y + vy*dt
    ax2, ay2 = aceleracao(x1, y1, vx1, vy1)
    vx2 = vx + ax2*dt
    vy2 = vy + ay2*dt
    x2 = x + vx1*dt
    y2 = y + vy1*dt
    x_novo = (x1 + x2)/2
    y_novo = (y1 + y2)/2
    vx_novo = (vx1 + vx2)/2
    vy_novo = (vy1 + vy2)/2
    return x_novo, y_novo, vx_novo, vy_novo

def runge_kutta(x, y, vx, vy):
    """
    Atualiza a posição e a velocidade do projétil com o método de 
    Runge-Kutta de 4ª ordem.
    """
    ax1, ay1 = aceleracao(x, y, vx, vy)  
    k1vx = ax1*dt
    k1vy = ay1*dt
    k1x = vx*dt
    k1y = vy*dt
    ax2, ay2 = aceleracao(x+k1x/2, y+k1y/2, vx+k1vx/2, vy+k1vy/2)  
    k2vx = ax2*dt
    k2vy = ay2*dt
    k2x = (vx+k1vx/2)*dt
    k2y = (vy+k1vy/2)*dt
    ax3, ay3 = aceleracao(x+k2x/2, y+k2y/2, vx+k2vx/2, vy+k2vy/2)  
    k3vx = ax3*dt
    k3vy = ay3*dt
    k3x = (vx+k2vx/2)*dt
    k3y = (vy+k2vy/2)*dt
    ax4, ay4 = aceleracao(x+k3x, y+k3y, vx+k3vx, vy+k3vy)  
    k4vx = ax4*dt
    k4vy = ay4*dt
    k4x = (vx+k3vx)*dt
    k4y = (vy+k3vy)*dt
    x_novo = x + 1/6*(k1x + 2*k2x + 2*k3x + k4x) 
    y_novo = y + 1/6*(k1y + 2*k2y + 2*k3y + k4y)
    vx_novo = vx + 1/6*(k1vx + 2*k2vx + 2*k3vx + k4vx)
    vy_novo = vy + 1/6*(k1vy + 2*k2vy + 2*k3vy + k4vy)
    return x_novo, y_novo, vx_novo, vy_novo
```

```python
def trajetoria(metodo, x0, y0, vx0, vy0, t0, t_max, y_min):
    """
    Função para calcular as coordenadas e as componentes da velocidade do projétil
    """
    # Usando as condições iniciais
    t = t0
    x = x0
    y = y0
    vx = vx0
    vy = vy0

    # Criando listas para armazenar os valores calculados
    t_salvo = np.array([t])
    x_salvo = np.array([x])
    y_salvo = np.array([y])
    vx_salvo = np.array([vx])
    vy_salvo = np.array([vy])

    # Atualizando os valores das variáveis
    while t < t_max:
        t = t + dt
        x, y, vx, vy = metodo(x, y, vx, vy)
        t_salvo = np.append(t_salvo, t)
        x_salvo = np.append(x_salvo, x)
        y_salvo = np.append(y_salvo, y)
        vx_salvo = np.append(vx_salvo, vx)
        vy_salvo = np.append(vy_salvo, vy)  
        if y < y_min:
            break
    
    return t_salvo, x_salvo, y_salvo, vx_salvo, vy_salvo
```

```python
# Constantes físicas
g = 9.81 # Aceleração da gravidade em m/s^2
```

```python
# Condições iniciais
t0 = 0
x0 = 0
y0 = 0
v0 = 1640
theta0 = 45
theta0_rad = np.deg2rad(theta0)
vx0 = v0*np.cos(theta0_rad)
vy0 = v0*np.sin(theta0_rad)

# Condições de parada
t_max = 10*60
y_min = 0 

# Incremento no tempo
dt = 10

# Método numérico
metodo = heun
```

```python
# Desenhando a trajetória do projétil para diferentes valores de k (símbolos)
# e comparando com a trajetória exata (linhas)
fig,ax = plt.subplots()
ax.figure.set_size_inches(12, 8)

k = 1e-8
t, x, y, vx, vy = trajetoria(metodo, x0, y0, vx0, vy0, t0, t_max, y_min)
ax.plot(x, y, 'o', markersize=10, markeredgecolor='k', markerfacecolor="None")
x = np.linspace(x0, max(x), 100)
y = func_analitica(x)
ax.plot(x, y, 'k--', linewidth=2, label='k = 0')

k = 1e-3
t, x, y, vx, vy = trajetoria(metodo, x0, y0, vx0, vy0, t0, t_max, y_min)
ax.plot(x, y, 'o', markersize=10, markeredgecolor='r', markerfacecolor="None")
x = np.linspace(x0, max(x), 100)
y = func_analitica(x)
ax.plot(x, y, 'r--', linewidth=2, label='k = %.1E'%k)

k = 5e-3
t, x, y, vx, vy = trajetoria(metodo, x0, y0, vx0, vy0, t0, t_max, y_min)
ax.plot(x, y, 'o', markersize=10, markeredgecolor='b', markerfacecolor="None")
x = np.linspace(x0, max(x), 100)
y = func_analitica(x)
ax.plot(x, y, 'b--', linewidth=2, label='k = %.1E'%k)

k = 1e-2
t, x, y, vx, vy = trajetoria(metodo, x0, y0, vx0, vy0, t0, t_max, y_min)
ax.plot(x, y, 'o', markersize=10, markeredgecolor='g', markerfacecolor="None")
x = np.linspace(x0, max(x), 100)
y = func_analitica(x)
ax.plot(x, y, 'g--', linewidth=2, label='k = %.1E'%k)

k = 2e-2
t, x, y, vx, vy = trajetoria(metodo, x0, y0, vx0, vy0, t0, t_max, y_min)
ax.plot(x, y, 'o', markersize=10, markeredgecolor='m', markerfacecolor="None")
x = np.linspace(x0, max(x), 100)
y = func_analitica(x)
ax.plot(x, y, 'm--', linewidth=2, label='k = %.1E'%k)

ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/1e3)))
ax.set_xlabel('$x$ (km)', size=28)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{0:g}'.format(y/1e3)))
ax.set_ylabel('$y$ (km)', size=28)
ax.set_ylim(bottom=0)
ax.tick_params(labelsize=22)
ax.legend(fontsize=16)
ax.grid()
fig.savefig("Resultado2.png", dpi=300, facecolor='white', bbox_inches='tight')
```

Vemos que, para valores muito pequenos de $k$ a trajetória se parece com uma parábola, como já esperávamos para o caso de resistência do ar muito pequena. Notamos que, quanto maior o valor de $k$, tanto o alcance quanto a altura máxima diminuem. Para valores grandes de $k$, o projétil praticamente cai na vertical.
